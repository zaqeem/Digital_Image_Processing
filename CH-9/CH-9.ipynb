```python
import numpy as np
import cv2

def thinning_opencv(img, kernel_size=3, iterations=1):
    # Create a structuring element (kernel) for morphological operations
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (kernel_size, kernel_size))
    
    # Perform the thinning (skeletonization) operation
    thinned_img = img.copy()
    
    for _ in range(iterations):
        eroded = cv2.erode(thinned_img, kernel)
        dilated = cv2.dilate(eroded, kernel)
        thinned_img = cv2.subtract(dilated, thinned_img)
    
    return thinned_img

# Load the image in grayscale mode
img = cv2.imread('FigP0905(U).tif', 0)

# Perform thinning with OpenCV's morphological operations
thinned_img = thinning_opencv(img, kernel_size=3, iterations=1)

# Display the results
cv2.imshow('Original Image', img)
cv2.imshow('Thinned Image', thinned_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import numpy as np
import cv2

def hm_transformation(image, se):
    img_rows, img_cols = image.shape
    se_rows, se_cols = se.shape

    pad_rows, pad_cols = (se_rows - 1) // 2, (se_cols - 1) // 2
    padded_image = np.pad(image, ((pad_rows, pad_rows), (pad_cols, pad_cols)), mode='constant', constant_values=0)

    hmt_image = np.zeros_like(image)

    for row in range(img_rows):
        for col in range(img_cols):
            roi = padded_image[row: row + se_rows, col: col + se_cols]
            masked_se = se.copy()
            masked_se[masked_se == -1] = roi[masked_se == -1]

            if np.all(roi[masked_se == 1] == 1):
                hmt_image[row, col] = 1
            else:
                hmt_image[row, col] = 0
    return hmt_image

def thinning(image, se_s):
    # Binarize the image
    image = image > 0
    pre = np.zeros_like(image)
    curr = image.copy()
    while np.any(curr != pre):
        pre = curr.copy()
        for se in se_s:
            hmt = hm_transformation(curr, se)
            hmt_comp = np.logical_not(hmt).astype(np.uint8)
            curr = np.logical_and(curr, hmt_comp).astype(np.uint8)
    return (curr * 255).astype(np.uint8)

# Load the image in grayscale mode
original_image = cv2.imread('FigP0905(U).tif', cv2.IMREAD_GRAYSCALE)

# Define the structuring elements
se1 = np.array([[0,0,0],[-1,1,-1],[1,1,1]])
se2 = np.array([[-1,0,0],[1,1,0],[1,1,-1]])
se3 = np.array([[1,-1,0],[1,1,0],[1,-1,0]])
se4 = np.array([[1,1,-1],[1,1,0],[-1,0,0]])
se5 = np.array([[1,1,1],[-1,1,-1],[0,0,0]])
se6 = np.array([[-1,1,1],[0,1,1],[0,0,-1]])
se7 = np.array([[0,-1,1],[0,1,1],[0,-1,1]])
se8 = np.array([[0,0,-1],[0,1,1],[-1,1,1]])

se_s = [se1, se2, se3, se4, se5, se6, se7, se8]

# Perform thinning
thinned_image = thinning(original_image, se_s)

# Concatenate original and thinned images side by side
concatenated_image = np.hstack((original_image, thinned_image))

# Display the concatenated image
cv2.imshow('Original and Thinned Image', concatenated_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def fill_holes(binary_image):
    # Step 1: Compute the complement of the binary image
    complement_image = cv2.bitwise_not(binary_image)
    
    # Step 2: Create the marker image F
    marker_image = np.zeros_like(binary_image)
    marker_image[:, 0] = 1 - binary_image[:, 0]
    marker_image[:, -1] = 1 - binary_image[:, -1]
    marker_image[0, :] = 1 - binary_image[0, :]
    marker_image[-1, :] = 1 - binary_image[-1, :]
    
    # Step 3: Perform morphological reconstruction by dilation
    prev_marker = marker_image.copy()
    while True:
        # Dilate marker image
        dilated_marker = cv2.dilate(prev_marker, np.ones((3, 3), np.uint8))
        # Constrain the dilated marker to the complement of the binary image
        new_marker = np.minimum(dilated_marker, complement_image)
        
        if np.array_equal(new_marker, prev_marker):
            break
        
        prev_marker = new_marker.copy()
    
    # Step 4: Complement the result to get the filled image
    filled_image = cv2.bitwise_not(new_marker)
    
    return filled_image

# Example usage
binary_image = cv2.imread('example_binary_image.png', cv2.IMREAD_GRAYSCALE)
binary_image = cv2.threshold(binary_image, 127, 255, cv2.THRESH_BINARY)[1]

filled_image = fill_holes(binary_image)

# Displaying the results
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.title('Original Binary Image')
plt.imshow(binary_image, cmap='gray')

plt.subplot(1, 2, 2)
plt.title('Filled Image')
plt.imshow(filled_image, cmap='gray')

plt.show()
```

```python
def otsu_threshold(image):
    hist, bins = np.histogram(image.flatten(), 256, [0,256])
    total_pixels = image.size

    current_max, threshold = 0, 0
    sum_total, sum_background, weight_background = 0, 0, 0

    for i in range(256):
        sum_total += i * hist[i]
    
    for i in range(256):
        weight_background += hist[i]
        if weight_background == 0:
            continue
        weight_foreground = total_pixels - weight_background
        if weight_foreground == 0:
            break
        
        sum_background += i * hist[i]
        mean_background = sum_background / weight_background
        mean_foreground = (sum_total - sum_background) / weight_foreground

        between_class_variance = weight_background * weight_foreground * (mean_background - mean_foreground) ** 2

        if between_class_variance > current_max:
            current_max = between_class_variance
            threshold = i

    return threshold
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image_path = 'Fig0940(a)(rice_image_with_intensity_gradient).tif'
img = cv2.imread(image_path, 0)

otsu_threshold_value = otsu_threshold(img)
threshold_img = np.where(img > otsu_threshold_value, 255, 0).astype('uint8')


se = np.ones((81, 81), dtype='uint8')


def erode(image, se):
    se_height, se_width = se.shape
    pad_height = se_height // 2
    pad_width = se_width // 2
    
    padded_image = np.pad(image, ((pad_height, pad_height), (pad_width, pad_width)), mode='constant', constant_values=255)
    result = np.zeros_like(image)
    
    for i in range(pad_height, padded_image.shape[0] - pad_height):
        for j in range(pad_width, padded_image.shape[1] - pad_width):
            region = padded_image[i - pad_height:i + pad_height + 1, j - pad_width:j + pad_width + 1]
            result[i - pad_height, j - pad_width] = np.min(region[se == 1])
    
    return result

def dilate(image, se):
    se_height, se_width = se.shape
    pad_height = se_height // 2
    pad_width = se_width // 2
    
    padded_image = np.pad(image, ((pad_height, pad_height), (pad_width, pad_width)), mode='constant', constant_values=0)
    result = np.zeros_like(image)
    
    for i in range(pad_height, padded_image.shape[0] - pad_height):
        for j in range(pad_width, padded_image.shape[1] - pad_width):
            region = padded_image[i - pad_height:i + pad_height + 1, j - pad_width:j + pad_width + 1]
            result[i - pad_height, j - pad_width] = np.max(region[se == 1])
    
    return result

def opening(image, se):
    eroded_img = erode(image, se)
    opened_img = dilate(eroded_img, se)
    return opened_img


opening_img = opening(img, se).astype('uint8')


top_hat_img = (img - opening_img).astype('uint8')


otsu_threshold_value_tophat = otsu_threshold(top_hat_img)
thresholded_top_hat_img = np.where(top_hat_img > otsu_threshold_value_tophat, 255, 0).astype('uint8')

plt.figure(figsize=(12, 6))

plt.subplot(231)
plt.imshow(img, cmap='gray')
plt.title('Original Image')

plt.subplot(232)
plt.imshow(threshold_img, cmap='gray')
plt.title('Otsu Thresholded Image')

plt.subplot(233)
plt.imshow(opening_img, cmap='gray')
plt.title('Opened Image with Disk SE')

plt.subplot(234)
plt.imshow(top_hat_img, cmap='gray')
plt.title('Top-Hat Transformed Image')

plt.subplot(235)
plt.imshow(thresholded_top_hat_img, cmap='gray')
plt.title('Otsu Thresholded Top-Hat Image')

plt.tight_layout()
plt.show()
```

```python
import numpy as np 
import cv2

def generate_disk(radius):
    diameter = 2 * radius + 1
    disk = np.ones((diameter, diameter), dtype=np.uint8)
    center = radius
    
    for y in range(diameter):
        for x in range(diameter):
            if (x - center)**2 + (y - center)**2 > radius**2:
                disk[y, x] = 0
    
    return disk
img=cv2.imread('Fig0943(a)(dark_blobs_on_light_background).tif',0)
structuring_element_30 = generate_disk(30)
closing_img=closing(img,structuring_element_30)

cv2.imshow('img',img)
cv2.imshow('closing',closing_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
print(structuring_element_30.max())
```
