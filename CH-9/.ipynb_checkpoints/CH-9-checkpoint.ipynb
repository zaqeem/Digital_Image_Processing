```python
import numpy as np
import cv2

def hm_transformation(image, se):
    img_rows, img_cols = image.shape
    se_rows, se_cols = se.shape

    pad_rows, pad_cols = (se_rows - 1) // 2, (se_cols - 1) // 2
    padded_image = np.pad(image, (pad_rows, pad_cols), mode='constant', constant_values=0)

    hmt_image = np.zeros_like(image)

    for row in range(img_rows):
        for col in range(img_cols):
            roi = padded_image[row: row + se_rows, col: col + se_cols]
            masked_se = se.copy()
            masked_se[masked_se == -1] = roi[masked_se == -1]

            if np.all(roi[masked_se == 1] == 1):
                hmt_image[row, col] = 1
            else:
                hmt_image[row, col] = 0
    return hmt_image

def thinning(image, se_s):
    image = image / 255
    pre = np.zeros_like(image)
    curr = image
    thinned_image = np.zeros_like(image)
    while np.any((curr != pre)):
        pre = curr
        for i in range(len(se_s)):
            hmt = hm_transformation(image, se_s[i])
            hmt_comp = np.logical_not(hmt) * 1
            curr = np.logical_and(image, hmt_comp) * 1
    thinned_image = curr
    return (thinned_image * 255).astype(np.uint8)

original_image = cv2.imread('FigP0905(U).tif', cv2.IMREAD_GRAYSCALE)
se1 = np.array([[0,0,0],[-1,1,-1],[1,1,1]])
se2 = np.array([[-1,0,0],[1,1,0],[1,1,-1]])
se3 = np.array([[1,-1,0],[1,1,0],[1,-1,0]])
se4 = np.array([[1,1,-1],[1,1,0],[-1,0,0]])
se5 = np.array([[1,1,1],[-1,1,-1],[0,0,0]])
se6 = np.array([[-1,1,1],[0,1,1],[0,0,-1]])
se7 = np.array([[0,-1,1],[0,1,1],[0,-1,1]])
se8 = np.array([[0,0,-1],[0,1,1],[-1,1,1]])

se_s = [se1, se2, se3, se4, se5, se6, se7, se8]
thinned_image = thinning(original_image, se_s)

# Concatenate original and thinned images side by side
concatenated_image = cv2.hconcat([original_image, thinned_image])

# Display the concatenated image
cv2.imshow('Original and Thinned Image', concatenated_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
from skimage.morphology import reconstruction
import matplotlib.pyplot as plt

# Step 1: Load the original image
image_path = 'Fig0931(a)(text_image).tif'
original_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Step 2: Erode the image with a structuring element of size 51x1
structuring_element = cv2.getStructuringElement(cv2.MORPH_RECT, (1, 51))
eroded_image = cv2.erode(original_image, structuring_element)

# Step 3: Perform standard opening (erosion followed by dilation)
opened_image = cv2.dilate(eroded_image, structuring_element)

# Step 4: Perform opening by reconstruction
# Use a 3x3 structuring element for reconstruction
marker = eroded_image
mask = original_image
reconstructed_image = reconstruction(marker, mask, method='dilation')

# Convert the reconstructed image from float64 to uint8
reconstructed_image = (reconstructed_image * 255).astype(np.uint8)

# Display the results
fig, axes = plt.subplots(1, 4, figsize=(20, 5))
axes[0].imshow(original_image, cmap='gray')
axes[0].set_title('Original Image (a)')
axes[0].axis('off')

axes[1].imshow(eroded_image, cmap='gray')
axes[1].set_title('Eroded Image (b)')
axes[1].axis('off')

axes[2].imshow(opened_image, cmap='gray')
axes[2].set_title('Opening (c)')
axes[2].axis('off')

axes[3].imshow(reconstructed_image, cmap='gray')
axes[3].set_title('Opening by Reconstruction (d)')
axes[3].axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def fill_holes(binary_image):
    # Step 1: Compute the complement of the binary image
    complement_image = cv2.bitwise_not(binary_image)
    
    # Step 2: Create the marker image F
    marker_image = np.zeros_like(binary_image)
    marker_image[:, 0] = 1 - binary_image[:, 0]
    marker_image[:, -1] = 1 - binary_image[:, -1]
    marker_image[0, :] = 1 - binary_image[0, :]
    marker_image[-1, :] = 1 - binary_image[-1, :]
    
    
    prev_marker = marker_image.copy()
    while True:
        dilated_marker = cv2.dilate(prev_marker, np.ones((3, 3), np.uint8))
        new_marker = np.minimum(dilated_marker, complement_image)
        
        if np.array_equal(new_marker, prev_marker):
            break
        
        prev_marker = new_marker.copy()
    
    filled_image = cv2.bitwise_not(new_marker)
    
    return filled_image

# Generate a synthetic binary image with a hole
binary_image = cv2.imread('Fig0931(a)(text_image).tif',0)
binary_image=binary_image/255
# Apply the hole-filling algorithm
filled_image = fill_holes(binary_image)

# Displaying the results
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.title('Original Binary Image')
plt.imshow(binary_image, cmap='gray')

plt.subplot(1, 2, 2)
plt.title('Filled Image')
plt.imshow(filled_image, cmap='gray')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def fill_holes(binary_image):
    # Step 1: Compute the complement of the binary image
    complement_image = cv2.bitwise_not(binary_image)
    
    # Step 2: Create the marker image F
    marker_image = np.zeros_like(binary_image)
    marker_image[:, 0] = 1 - binary_image[:, 0]
    marker_image[:, -1] = 1 - binary_image[:, -1]
    marker_image[0, :] = 1 - binary_image[0, :]
    marker_image[-1, :] = 1 - binary_image[-1, :]
    
    # Step 3: Perform morphological reconstruction by dilation
    prev_marker = marker_image.copy()
    while True:
        # Dilate marker image
        dilated_marker = cv2.dilate(prev_marker, np.ones((3, 3), np.uint8))
        # Constrain the dilated marker to the complement of the binary image
        new_marker = np.minimum(dilated_marker, complement_image)
        
        if np.array_equal(new_marker, prev_marker):
            break
        
        prev_marker = new_marker.copy()
    
    # Step 4: Complement the result to get the filled image
    filled_image = cv2.bitwise_not(new_marker)
    
    return filled_image

# Example usage
binary_image = cv2.imread('example_binary_image.png', cv2.IMREAD_GRAYSCALE)
binary_image = cv2.threshold(binary_image, 127, 255, cv2.THRESH_BINARY)[1]

filled_image = fill_holes(binary_image)

# Displaying the results
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.title('Original Binary Image')
plt.imshow(binary_image, cmap='gray')

plt.subplot(1, 2, 2)
plt.title('Filled Image')
plt.imshow(filled_image, cmap='gray')

plt.show()
```

```python
import cv2
import numpy as np
def erosion(img,kernel):
    r,c=img.shape[:2]
    row,col=kernel.shape[0]//2,kernel.shape[1]//2
    pad_img = np.pad(img,((row,row),(col, col)), mode='reflect')
    ero=np.zeros(img.shape)
    for i in range(r):
        for j in range(c):
            ero[i,j]=np.min(pad_img[i:i+row*2+1,j:j+col*2+1])
            
    return ero
img=cv2.imread('Fig0935(a)(ckt_board_section).tif',0)
se=np.ones((5,5),dtype='uint8')
eroted_img=erosion(img,se).astype('uint8')
cv2.imshow('img',img)
cv2.imshow('erosion',eroted_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
def dilation(img,kernel):
    r,c=img.shape[:2]
    row,col=kernel.shape[0]//2,kernel.shape[1]//2
    pad_img = np.pad(img,((row,row),(col, col)), mode='reflect')
    dilate=np.zeros(img.shape)
    for i in range(r):
        for j in range(c):
            roi=pad_img[i:i+row*2+1,j:j+col*2+1]
            dilate[i,j]=np.max(roi)
    return dilate
img=cv2.imread('Fig0935(a)(ckt_board_section).tif',0)
se=np.ones((5,5),dtype='uint8')
dilated_img=dilation(img,se).astype('uint8')
cv2.imshow('img',img)
cv2.imshow('dilated',dilated_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
def opening(img,se):
    eroted_img=erosion(img,se)
    opening_img=dilation(eroted_img,se)
    return opening_img
def closing(img,se):
    dilated_img=dilation(img,se)
    closing_img=erosion(dilated_img,se)
    return closing_img
img=cv2.imread('Fig0935(a)(ckt_board_section).tif',0)
se=np.ones((7,7),dtype='uint8')
Se=np.ones((11,11),dtype='uint8')
opening_img=opening(img,se).astype('uint8')
closing_img=closing(img,Se).astype('uint8')
cv2.imshow('img',img)
cv2.imshow('opening',opening_img)
cv2.imshow('closing',closing_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
